## this and super keywords
### super class is to call the parent method & constructor(everything non-private)

```python

class shape{
    private boolean filled;
    private String color;
    public shape()
    {
       color = "red";
    }

    public boolean getfilled(){
        return filled;
    }

    public void setcolor(String color){
        this.color = color;
    }
    public String getcolor(){
        return color;
    }
    public void type(int i){
        System.out.println("integer");
    }
}

class circle extends shape{
    private double radius;

    public circle(double radius,String color){
        this.radius = radius;
        super.setcolor(color);
    }
    public String toString() {
        return "Circle{" + "radius=" + radius + ", color=" + getcolor() + '}';
    }
    public void type(double i){
        System.out.println("double");
    }
}


circle c = new circle(2,"orange");
System.out.println(c.toString());
-------------------------------------
Circle{radius=2.0, color=orange}

用了super调用了shape里的setcolor method,变相地调用了父类的属性color
```
