### node增加
从 1->2->3 到 1->2->4->3  
原理为将“2”赋予一个reference type，将新加的node指向“3”，再将“2”指向新的node

```python
 public static void main(String[] args){
        Node<Integer> head = new Node<>(1);
        head.next = new Node<>(2);
        head.next.next = new Node<>(3);
        Node<Integer> addition = new Node<>(4);
        Node<Integer> O = head.next;
        addition.next = head.next.next;
        O.next = addition;
        System.out.println(head);
    }
class Node<E>{
    E element;
    Node<E> next;
    public Node(E e){
        element = e;
    }
    public String toString() {
        StringBuilder sb = new StringBuilder();
        Node<E> current = this;
        while (current != null) {
            sb.append(current.element);
            if (current.next != null) {
                sb.append(" -> ");
            }
            current = current.next;
        }
        return sb.toString();
    }
}
```
