### 1. 代价函数定义：均方误差 (MSE)

在回归问题中，我们通常使用均方误差（Mean Squared Error）作为代价函数。它衡量了模型预测值与真实值之间的平均平方误差。

公式如下：
$$J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)^2$$

其中：
* $m$ 是训练样本的数量。
* $f_{w,b}(x^{(i)}) = w x^{(i)} + b$ 是模型的预测值（以简单线性回归为例）。
* $y^{(i)}$ 是第 $i$ 个样本的真实标签。
* 前面的 $\frac{1}{2}$ 是为了在求导（梯度下降）时消去平方项带来的 $2$，使计算更简洁。
