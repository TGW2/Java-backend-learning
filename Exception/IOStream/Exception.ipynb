# Java 异常处理笔记

## 1. Unchecked Exceptions (非检查型异常)

**定义：** 包含 `RuntimeException`、`Error` 及其子类。

* **发生场景：** 通常由**编程逻辑错误**引起，可能在程序的任何地方发生。
* **编译器行为：** Java **不强制要求**编写代码来处理它们，以避免 `try-catch` 块的过度使用。
* **常见示例：**
    * **Error 的子类：**
        * `VirtualMachineError`: JVM 损坏或运行资源耗尽。
    * **RuntimeException 的子类：**
        * `ArithmeticException`: 算术错误，如**整数除以零**。
        * `IndexOutOfBoundsException`: **下标越界**（如访问数组非法索引）。
        * `NumberFormatException`: **字符串转换数值失败**（例如将 "abc" 解析为整数）。
        * `NullPointerException`: **空指针异常**（在对象赋值前调用其引用变量）。

---

## 2. Checked Exceptions (检查型异常)

**定义：** 除了 `RuntimeException` 以外的所有 `Exception` 子类。

* **编译器行为：** 编译器**强制要求**开发者处理这些异常。如果不对其进行捕获或声明抛出，程序将无法通过编译。
* **常见示例 (主要是 `IOException` 的子类)：**
    * `EOFException`: 尝试读取超过文件末尾的内容。
    * `FileNotFoundException`: 尝试打开或写入一个不存在的文件。
    * `MalformedURLException`: 使用了无效格式的 URL。
