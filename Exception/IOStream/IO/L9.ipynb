## InputStream & OutputStream


## **InputStream** Syntax 

```python
read():int //read next byte of data as int (0 to 255), or -1 if end of stream
read(b:byte[], off:int, len:int):int //read up to len bytes into byte array b starting at offset off, return number of bytes read or -1 if end of stream
available():int //return number of bytes that can be read without blocking
close():void //close the input stream and release any system resources associated with it
```

```python
import java.io.IOException;

public class InputStreamAllMethods {
    public static void main(String[] args) throws IOException {

        System.out.print("Type something: ");

        // 1. available()
        int available = System.in.available();
        System.out.println("\nAvailable before read: " + available);

        // 2. read() - read one byte
        int oneByte = System.in.read();
        System.out.println("read(): " + oneByte + " -> " + (char) oneByte);

        // 3. read(byte[], off, len)
        byte[] buf = new byte[10];
        int n = System.in.read(buf, 0, 5);
        System.out.println("read(byte[], off, len): bytes read = " + n);

        for (int i = 0; i < n; i++) {
            System.out.println("buf[" + i + "] = " + (char) buf[i]);
        }

        // 4. close()
        System.in.close();
    }
}
```

## 不用available更efficient的方法

```python
DataInputStream in = new DataInputStream(new FileInputStream("src/L10.txt"));
boolean EOF = false;
while (!EOF) {
    try {
        String a = in.readUTF();
        System.out.println(a);
    } catch (EOFException e) {
        EOF = true;
    }
}
in.close();
```
